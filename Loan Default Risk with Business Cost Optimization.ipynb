```python
# Business Cost Values

FP_COST = 1
FN_COST = 5

print("False Positive Cost :", FP_COST)
print("False Negative Cost :", FN_COST)
```

```python
from sklearn.metrics import confusion_matrix

def business_cost(y_true, y_pred, fp_cost, fn_cost):
    
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    
    total_cost = (fp * fp_cost) + (fn * fn_cost)
    
    return total_cost, tn, fp, fn, tp
```

```python
thresholds = np.arange(0.10, 0.91, 0.05)

results = []

for threshold in thresholds:
    
    predictions = (best_prob >= threshold).astype(int)
    
    cost, tn, fp, fn, tp = business_cost(
        y_test,
        predictions,
        FP_COST,
        FN_COST
    )
    
    results.append([
        threshold,
        cost,
        tn,
        fp,
        fn,
        tp
    ])
```

```python
cost_df = pd.DataFrame(results, columns=[
    "Threshold",
    "Business Cost",
    "TN",
    "FP",
    "FN",
    "TP"
])

cost_df
```

```python
best_result = cost_df.loc[cost_df["Business Cost"].idxmin()]

best_threshold = best_result["Threshold"]

print("Optimal Threshold :", best_threshold)
print("Minimum Business Cost :", best_result["Business Cost"])
```
